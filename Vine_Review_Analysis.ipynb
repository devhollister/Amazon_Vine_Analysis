```python
import pandas as pd
```

```python
#read in csv
df = pd.read_csv('vine_table')
df.head()
```

```python
#filter df
filtered_df = df[df["total_votes"] >=20.0]
filtered_df
```

```python
#filter by helpful/total votes >= 50%
filtered_2 = filtered_df[filtered_df['helpful_votes']/filtered_df['total_votes'] >= .50] 
filtered_2
```

```python
#filter by vine program yes
paid = filtered_2[filtered_2['vine']=='Y']
paid.head()
```

```python
#filter by vine program no
unpaid = filtered_2[filtered_2['vine']=='N']
unpaid.head()
```

```python
#total reviews paid
len(paid)
```

```python
#total reviews unpaid
len(unpaid)
```

```python
# 5 star ratings paid
len(paid[paid['star_rating']==5])
```

```python
# 5 star ratings unpaid
len(unpaid[unpaid['star_rating']==5])
```

```python
#Percent 5 star paid
len(paid[paid['star_rating']==5])/len(paid)*100
```

```python
# Percent 5 star unpaid
len(unpaid[unpaid['star_rating']==5])/len(unpaid)*100
```
